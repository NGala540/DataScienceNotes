**displot**: Ta funkcja służy do tworzenia wykresów rozkładu jednowymiarowych danych. Pozwala na generowanie różnych typów wykresów, takich jak histogramy, wykresy gęstości jądra (KDE) i dystrybuanty empiryczne. Dodatkowo, displot umożliwia tworzenie wykresów dla wielu zmiennych jednocześnie, co ułatwia porównywanie rozkładów. Możemy również dostosować wygląd wykresu za pomocą różnych parametrów.

**distplot**: Ta funkcja jest starszą wersją displot, która jest teraz przestarzała i zaleca się używanie displot zamiast niej. Distplot również służy do tworzenia wykresów rozkładu jednowymiarowych danych, ale ma mniej opcji i mniejszą elastyczność w porównaniu z displot.

**histplot**: Ta funkcja pozwala na tworzenie wykresów histogramów jedno- i dwuwymiarowych. Histogramy pozwalają na analizowanie rozkładu danych poprzez podział wartości na przedziały i prezentację liczby obserwacji w każdym z nich. Histplot umożliwia także tworzenie wykresów gęstości jądra (KDE) oraz przedstawienie danych w postaci wykresów słupkowych (bar plot) dla danych kategorycznych.

```python
palette = sns.color_palette("tab10", 3)

fig, [ax1, ax2] = plt.subplots(nrows = 2, ncols = 1, figsize = (16,8), sharex = True)

sns.histplot(wine['Alcohol'], ax = ax1, color = palette[0], binwidth = 0.5, alpha = 0.5, )
sns.histplot(wine['Malic acid'], ax = ax1, color = palette[1], binwidth = 0.5, alpha = 0.5)
sns.histplot(wine['Ash'], ax = ax1, color = palette[2], binwidth = 0.5, alpha = 0.5)

sns.boxplot(data = wine[['Alcohol', 'Malic acid', 'Ash']],  ax = ax2, palette = palette, orient = 'h')

ax1.legend(['Alcohol', 'Malic acid', 'Ash']);
```

**stat:**

* count: Wartość parametru stat jako "count" powoduje, że wysokość słupków na histogramie przedstawia liczbę obserwacji przypadających na dany przedział (kosz). To jest domyślne zachowanie dla histogramów w Seaborn.
* frequency: Wartość parametru stat jako "frequency" sprawia, że wysokość słupków na histogramie reprezentuje liczbę obserwacji w danym przedziale, podzieloną przez szerokość tego przedziału, co pozwala na analizę danych z uwzględnieniem różnych szerokości przedziałów.
* density: Ustawienie parametru stat na "density" sprawia, że wysokość słupków na histogramie reprezentuje gęstość prawdopodobieństwa w danym przedziale, normalizując dane tak, że całkowite pole pod histogramem wynosi 1. Dzięki temu histogram staje się estymatorem gęstości prawdopodobieństwa.
* probability: Parametr stat ustawiony na "probability" powoduje, że wysokość słupków na histogramie przedstawia prawdopodobieństwo wystąpienia obserwacji w danym przedziale, normalizując dane tak, że suma wysokości słupków na wykresie wynosi 1. Umożliwia to porównywanie histogramów z różnymi liczbami obserwacji.


**bins:** Parametr ten określa liczbę przedziałów (koszy) używanych do podziału danych na histogramie. Większa liczba koszy skutkuje większą precyzją wykresu, ale może prowadzić do szumów i trudności w interpretacji danych. Mniejsza liczba koszy może natomiast prowadzić do utraty istotnych informacji.

**binwidth:** Ten parametr pozwala na określenie szerokości każdego przedziału (kosza) na histogramie. Alternatywnie, zamiast określać liczbę koszy, możemy określić ich szerokość, co może być przydatne w przypadku analizy danych o konkretnej skali. Podanie wartości dla binwidth spowoduje, że parametr bins zostanie zignorowany.

**binrange:** Parametr ten umożliwia określenie zakresu wartości na histogramie, czyli zakresu danych, które mają być uwzględnione w analizie. Jest to szczególnie przydatne, gdy chcemy skupić się na określonym obszarze danych, ignorując wartości ekstremalne lub szum. binrange przyjmuje krotkę (tuple) z dwiema wartościami, które określają dolną i górną granicę zakresu.

```python
palette = sns.color_palette("tab10", 3)

fig, ax1 = plt.subplots(nrows = 1, ncols = 1, figsize = (16,8), sharex = True)

sns.histplot(wine['Alcohol'], ax = ax1, color = palette[0], binwidth = 0.5, alpha = 0.5, stat = 'probability')
sns.histplot(wine['Malic acid'], ax = ax1, color = palette[1], binwidth = 0.5, alpha = 0.5, stat = 'probability')
sns.histplot(wine['Ash'], ax = ax1, color = palette[2], binwidth = 0.5, alpha = 0.5, stat = 'probability')
```

```python
# Stwórz obiekt fig, ax używając funkcji subplots()
fig, ax = plt.subplots(3, 2, figsize=(14, 18))

# Magnesium
sns.boxplot(data=wine, x='Class', y='Magnesium', ax=ax[0, 0])
sns.histplot(data=wine, x='Magnesium', hue='Class', ax=ax[0, 1], stat = 'probability', bins = 25)

# Color intensity
sns.boxplot(data=wine, x='Class', y='Color intensity', ax=ax[1, 0])
sns.histplot(data=wine, x='Color intensity', hue='Class',  ax=ax[1, 1], stat = 'probability', bins = 25)

# Flavanoids
sns.boxplot(data=wine, x='Class', y='Flavanoids', ax=ax[2, 0])
sns.histplot(data=wine, x='Flavanoids', hue='Class', ax=ax[2, 1], stat = 'probability', bins = 25)

# Dodaj tytuły do wykresów
ax[0, 0].set_title('Boxplot: Magnesium')
ax[0, 1].set_title('Histplot: Magnesium')
ax[1, 0].set_title('Boxplot: Color intensity')
ax[1, 1].set_title('Histplot: Color intensity')
ax[2, 0].set_title('Boxplot: Flavanoids')
ax[2, 1].set_title('Histplot: Flavanoids')

# Wyświetl wykresy
plt.tight_layout()
plt.show()
```

## Wykres skrzypcowy

Kolejną zaletą Seaborn jest to, że ma gotowe motywy. Domyślny motyw nazywa się „darkgrid”.

Następnie zmienimy motyw na "whitegrid” podczas tworzenia wykresu skrzypcowego (violinplot), pokazującego również rozkład wartości (inaczej niż w boxplotach).

```python
sns.set_style('ticks' )

sns.violinplot(x='Class', y='Alcohol', data=wine, )

plt.grid()
```

## Rozkład gęstości - Kdeplot

```python
f(x) = (1/n * h) * Σ K((x - x_i) / h)

n to liczba próbek
x_i to wartość i-tej próbki
h to szerokość okna (szerokość jądra), która wpływa na wygładzenie estymacji
K to funkcja jądra (kernel), która opisuje kształt lokalnego wygładzenia
```

```python
sns.kdeplot(x = wine['Alcohol'], y = wine['Color intensity'])
plt.grid()
```

## Rozkład dytrybucji - jointplot

```python
typy = ["scatter" , "kde" , "hist" , "hex" , "reg" , "resid" ]
```

```python
for typ in typy:
    
    try:
        sns.jointplot(x='Alcohol', y='Color intensity', data = wine, hue ='Class', kind = typ)
        plt.show()
    
    except ValueError:
        print(f'Nie obsługiwany w tej formie typ wykresu: {typ}')
```
