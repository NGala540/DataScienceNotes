```python
# Zadanie 3.
# Dodaj do ramki trzeciÄ… kolumnÄ™, w ktÃ³rej znajdzie siÄ™ procentowy wynik uczniÃ³w i uczennic z tego egzaminu (to znaczy, Å¼e
# kaÅ¼da wartoÅ›Ä‡ liczbowa musi zostaÄ‡ podzielona przez maksimum punktÃ³w, jakie moÅ¼na byÅ‚o na tym egzaminie uzyskaÄ‡)

df_uczniowie['wynik_procentowy'] = df_uczniowie.wynik.apply(lambda x: x/100)

# NastÄ™pnie, chcemy zanonimizowaÄ‡ kolumnÄ™ z imionami: chcemy, by zostaÅ‚a tylko pierwsza i ostatnia litera imienia. Np. dla imienia â€œMarcelinaâ€ chcemy zachowaÄ‡ â€œM...aâ€

df_uczniowie['uczen_ukryte'] = df_uczniowie.uczen.apply(lambda x: f"{x[0]}...{x[len(x)-1]}") # lub x[-1] zamiast x[len(x)-1]
```

```python
# Zadanie 5.

# PoÅ‚Ä…cz ze sobÄ… dwie poniÅ¼sze ramki (doklej jednÄ…Â do drugiej)

student_data1 = pd.DataFrame({
    'student_id': ['S1', 'S2', 'S3', 'S4', 'S5'],
    'name': ['Danniella Fenton', 'Ryder Storey', 'Bryce Jensen', 'Ed Bernal', 'Kwame Morin']
    })
 
student_data2 = pd.DataFrame({
    'student_id': ['S6', 'S7', 'S8', 'S9', 'S10'],
    'name': ['Scarlette Fisher', 'Carla Williamson', 'Dante Morse', 'Kaiser William', 'Madeeha Preston']
    })

student_data = student_data1.append(student_data2) # wkrÃ³tce metoda frame.append bÄ™dzie usuniÄ™ta, wiÄ™c nie warto z niej korzystaÄ‡

student_data = pd.concat([student_data1, student_data2], axis=0)

student_data = student_data1.merge(student_data2, how='outer') # dodatkowo resetuje za nas indeks po zÅ‚Ä…czeniu (w porÃ³wnaniu do pd.concat(), ktÃ³ry tego nie robi)
student_data
```

```python
# Zadanie 6.

# Do ramki utworzonej w poprzednim zadaniu dodaj informacje o rezultatach studentÃ³w z ramki poniÅ¼ej

exam_results = pd.DataFrame({
    'student_id': ['S10', 'S3', 'S1', 'S7', 'S9', 'S5', 'S4', 'S8', 'S6', 'S13'],
    'marks': [210, 190, 222, 199, 201, 200, 198, 219, 201, 250]
    }) # zmodyfikowaliÅ›my exam_results usuwajÄ…c jednego naszego studenta i dodajÄ…c jednego obcego

student_data_with_exam = student_data.merge(exam_results, how='outer', on='student_id') # moÅ¼emy zaobserwowaÄ‡ jak wpÅ‚ywajÄ… rÃ³Å¼ne wartoÅ›ci argumentu 'how'
student_data_with_exam
```
