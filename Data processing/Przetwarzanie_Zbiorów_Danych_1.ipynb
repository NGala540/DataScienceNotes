```python
# Zadanie 3.
# Dodaj do ramki trzecią kolumnę, w której znajdzie się procentowy wynik uczniów i uczennic z tego egzaminu (to znaczy, że
# każda wartość liczbowa musi zostać podzielona przez maksimum punktów, jakie można było na tym egzaminie uzyskać)

df_uczniowie['wynik_procentowy'] = df_uczniowie.wynik.apply(lambda x: x/100)

# Następnie, chcemy zanonimizować kolumnę z imionami: chcemy, by została tylko pierwsza i ostatnia litera imienia. Np. dla imienia “Marcelina” chcemy zachować “M...a”

df_uczniowie['uczen_ukryte'] = df_uczniowie.uczen.apply(lambda x: f"{x[0]}...{x[len(x)-1]}") # lub x[-1] zamiast x[len(x)-1]
```

```python
# Zadanie 5.

# Połącz ze sobą dwie poniższe ramki (doklej jedną do drugiej)

student_data1 = pd.DataFrame({
    'student_id': ['S1', 'S2', 'S3', 'S4', 'S5'],
    'name': ['Danniella Fenton', 'Ryder Storey', 'Bryce Jensen', 'Ed Bernal', 'Kwame Morin']
    })
 
student_data2 = pd.DataFrame({
    'student_id': ['S6', 'S7', 'S8', 'S9', 'S10'],
    'name': ['Scarlette Fisher', 'Carla Williamson', 'Dante Morse', 'Kaiser William', 'Madeeha Preston']
    })

student_data = student_data1.append(student_data2) # wkrótce metoda frame.append będzie usunięta, więc nie warto z niej korzystać

student_data = pd.concat([student_data1, student_data2], axis=0)

student_data = student_data1.merge(student_data2, how='outer') # dodatkowo resetuje za nas indeks po złączeniu (w porównaniu do pd.concat(), który tego nie robi)
student_data
```

```python
# Zadanie 6.

# Do ramki utworzonej w poprzednim zadaniu dodaj informacje o rezultatach studentów z ramki poniżej

exam_results = pd.DataFrame({
    'student_id': ['S10', 'S3', 'S1', 'S7', 'S9', 'S5', 'S4', 'S8', 'S6', 'S13'],
    'marks': [210, 190, 222, 199, 201, 200, 198, 219, 201, 250]
    }) # zmodyfikowaliśmy exam_results usuwając jednego naszego studenta i dodając jednego obcego

student_data_with_exam = student_data.merge(exam_results, how='outer', on='student_id') # możemy zaobserwować jak wpływają różne wartości argumentu 'how'
student_data_with_exam
```
